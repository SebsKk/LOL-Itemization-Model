```python
# defining a function to load our data from json 

import json

def load_json(json_file):
    with open(json_file, 'r') as f:
        data = json.load(f)

    return data
```

```python
# loading and merging all the json files 


data = []

json_files = ['players_match_data_one.json.json','players_match_data_two.json.json','players_match_data_three.json','players_match_data_four.json','players_match_data_five.json', 'players_match_data_six.json','players_match_data_eight.json','players_match_data_nine.json','players_match_data_ten.json', 'players_match_data_eleven.json','players_match_data_twelve.json', 'players_match_data_thirteen.json']

for file in json_files:
    js = load_json(file)
    data.extend(js)
```

```python
# i now need to add 2 more keys to each dictionary representing player; champs played with and against. Every 10 elements represent 10 players from one match

def get_team_compositions (data):

   counter = 0
   holder = 0

   for i in range(len(data)):

      if counter == 10:
         counter = 0

      if counter == 0: 
         holder = i

      played_against_first_team = [[data[holder]['champion_name'], data[holder]['team_position']],[data[holder+1]['champion_name'],data[holder+1]['team_position']],[data[holder+2]['champion_name'],data[holder+2]['team_position']],[data[holder+3]['champion_name'],data[holder+3]['team_position']],[data[holder+4]['champion_name'],data[holder+4]['team_position']]]
      played_against_second_team = [[data[holder+5]['champion_name'], data[holder+5]['team_position']],[data[holder+6]['champion_name'],data[holder+6]['team_position']],[data[holder+7]['champion_name'],data[holder+7]['team_position']],[data[holder+8]['champion_name'],data[holder+8]['team_position']],[data[holder+9]['champion_name'],data[holder+9]['team_position']]]
      
      if counter <= 4:
         data[i]['played_against'] = played_against_second_team
         data[i]['played_with'] = [champ for champ in played_against_first_team if champ[0] != data[i]['champion_name']]

      elif counter > 4:
         data[i]['played_against'] = played_against_first_team
         data[i]['played_with'] = [champ for champ in played_against_second_team if champ[0] != data[i]['champion_name']]

      counter += 1 

   return data

    
```

```python
# assign the team comps

player_stats = get_team_compositions(data)
```

```python
# swapping champion names with those positions

def normalize_positions(player_stats):
    
    for i in range(len(player_stats)):
        for champ in player_stats[i]['played_against']:
            for key, value in champions.items():
                if champ[0] in value:
                    champ[0] = key

    for i in range(len(player_stats)):
        for champ in player_stats[i]['played_with']:
            for key, value in champions.items():
                if champ[0] in value:
                    champ[0] = key
    return player_stats

player_stats = normalize_positions(player_stats)
```

```python
# i need to break each nested column into different columns

def get_champion_class(champion):
    for class_name, champ_list in champions.items():
        if champion in champ_list:
            return class_name
    return 'Unknown'

def break_champions(df, champs_against, champs_with):

    for index, row in df.iterrows():

        for position in df.loc[index, champs_against]:
            if position[1] == 'JUNGLE':
                df.loc[index, 'Enemy Jungler'] = position[0]
            elif position[1] == 'MIDDLE':
                df.loc[index, 'Enemy Middle'] = position[0]
            elif position[1] == 'BOTTOM':
                df.loc[index, 'Enemy Bottom'] = position[0]
            elif position[1] == 'TOP':
                df.loc[index, 'Enemy Top'] = position[0]
            elif position[1] == 'UTILITY':
                df.loc[index, 'Enemy Utility'] = position[0]

    for index, row in df.iterrows():

        for position in df.loc[index, champs_with]:

            if position[1] == 'JUNGLE' or  df.loc[index,'team_position'] == 'JUNGLE':
                df.loc[index, 'Friendly Jungler'] = position[0]
            elif position[1] == 'MIDDLE' or df.loc[index,'team_position'] == 'MIDDLE':
                df.loc[index, 'Friendly Middle'] = position[0]
            elif position[1] == 'BOTTOM' or df.loc[index,'team_position'] == 'BOTTOM':
                df.loc[index, 'Friendly Bottom'] = position[0]
            elif position[1] == 'TOP' or df.loc[index,'team_position'] == 'TOP':
                df.loc[index, 'Friendly Top'] = position[0]
            elif position[1] == 'UTILITY' or df.loc[index,'team_position'] == 'UTILITY':
                df.loc[index, 'Friendly Utility'] = position[0]


    # Fill in the position of the champion played by the player

    position_mapping = {
        'JUNGLE': 'Jungler',
        'MIDDLE': 'Middle',
        'BOTTOM': 'Bottom',
        'TOP': 'Top',
        'UTILITY': 'Utility'
    }
    
    champion_classes = {champ: get_champion_class(champ) for champ in set(sum(champions.values(), []))}
    
    for index, row in df.iterrows():
        player_champ = row['champion_name']
        player_position = row['team_position']
        if player_position in position_mapping:
            column = f'Friendly {position_mapping[player_position]}'
            df.loc[index, column] = player_champ
            df.loc[index, f'{column} Class'] = champion_classes.get(player_champ, 'Unknown')

    return df
   
```

```python

# extract primary and secondary runes into separate columns

def extract_runes(df):
    def safe_extract(x, key):
        if isinstance(x, dict):
            return x.get(key, None)
        elif isinstance(x, str):
            try:
                return eval(x).get(key, None)
            except:
                return None
        else:
            return None
        
    df['Primary Rune'] = df['masteries'].apply(lambda x: safe_extract(x, 'primary_master'))
    df['Secondary Rune'] = df['masteries'].apply(lambda x: safe_extract(x, 'secondary_mastery'))
    
    return df


try:
    new_df = extract_runes(new_df)
    print("Runes extracted successfully.")
    print(new_df[['Primary Rune', 'Secondary Rune']].head())
except Exception as e:
    print(f"An error occurred: {e}")
    print("Sample of 'masteries' column:")
    print(new_df['masteries'].head())
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

def preprocess_lol_data_final_v2(df):
    # Columns to drop
    cols_to_drop = ['win', 'kills', 'deaths', 'assists']
    df = df.drop(cols_to_drop, axis=1)

    # Label encode champion_name    le = LabelEncoder()
    df['champion_id'] = le.fit_transform(df['champion_name'])

    # Columns to one-hot encode (excluding champion_name)
    categorical_cols = [
        'Enemy Top', 'Enemy Jungler', 'Enemy Middle', 'Enemy Bottom', 'Enemy Utility',
        'Friendly Jungler', 'Friendly Middle', 'Friendly Bottom', 'Friendly Utility', 'Friendly Top',
        'Primary Rune', 'Secondary Rune'
    ]
    
    # One-hot encoding for other categorical variables
    encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
    encoded = encoder.fit_transform(df[categorical_cols])
    encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(categorical_cols))
    
    def process_items(items):
        # Assuming items is a string representation of a list
        item_list = eval(items) if isinstance(items, str) else items
        return [item for item in item_list if item != 0]
    
    df['processed_items'] = df['items'].apply(process_items)
    
    champion_df = df[['champion_id']]
    other_features_df = encoded_df.copy()
    
    return champion_df, other_features_df, df['processed_items'], le

champions_df, other_features_df, items, le = preprocess_lol_data_final_v2(new_df)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import pandas as pd

def preprocess_items_multilabel(items_series):
    # Remove zeros and convert to list of lists
    items_list = [[item for item in items if item != 0] for items in items_series]
    
    # Use MultiLabelBinarizer to convert to binary format
    mlb = MultiLabelBinarizer()
    items_binary = mlb.fit_transform(items_list)
    
    # Create DataFrame with column names
    items_df = pd.DataFrame(items_binary, columns=[f'item_{item}' for item in mlb.classes_])
    
    return items_df, mlb

items_df, item_encoder = preprocess_items_multilabel(items)
```

```python
# now save it for the work in the next notebook
import pandas as pd
import pickle
import os

def save_processed_data(champions_df, other_features_df, items_df, item_encoder, le, base_dir='processed_data'):
    os.makedirs(base_dir, exist_ok=True)
    
    champions_df.to_csv(os.path.join(base_dir, 'champions_df.csv'), index=False)
    other_features_df.to_csv(os.path.join(base_dir, 'other_features_df.csv'), index=False)
    items_df.to_csv(os.path.join(base_dir, 'items_df.csv'), index=False)
    
    # Save encoders using pickle
    with open(os.path.join(base_dir, 'item_encoder.pkl'), 'wb') as f:
        pickle.dump(item_encoder, f)
    
    with open(os.path.join(base_dir, 'champion_encoder.pkl'), 'wb') as f:
        pickle.dump(le, f)
    
    print(f"All data saved in '{base_dir}' directory.")


save_processed_data(champions_df, other_features_df, items_df, item_encoder, le)
```
