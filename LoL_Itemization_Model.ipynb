```python
# defining a function to load our data from json 

import json

def load_json(json_file):
    with open(json_file, 'r') as f:
        data = json.load(f)

    return data
```

```python
# loading and merging all the json files 


data = []

json_files = ['players_match_data_one.json.json','players_match_data_two.json.json','players_match_data_three.json','players_match_data_four.json','players_match_data_five.json', 'players_match_data_six.json','players_match_data_eight.json','players_match_data_nine.json','players_match_data_ten.json', 'players_match_data_eleven.json','players_match_data_twelve.json', 'players_match_data_thirteen.json']

for file in json_files:
    js = load_json(file)
    data.extend(js)
```

```python
# i now need to add 2 more keys to each dictionary representing player; champs played with and against. Every 10 elements represent 10 players from one match

def get_team_compositions (data):

   counter = 0
   holder = 0

   for i in range(len(data)):

      if counter == 10:
         counter = 0

      if counter == 0: 
         holder = i

      played_against_first_team = [[data[holder]['champion_name'], data[holder]['team_position']],[data[holder+1]['champion_name'],data[holder+1]['team_position']],[data[holder+2]['champion_name'],data[holder+2]['team_position']],[data[holder+3]['champion_name'],data[holder+3]['team_position']],[data[holder+4]['champion_name'],data[holder+4]['team_position']]]
      played_against_second_team = [[data[holder+5]['champion_name'], data[holder+5]['team_position']],[data[holder+6]['champion_name'],data[holder+6]['team_position']],[data[holder+7]['champion_name'],data[holder+7]['team_position']],[data[holder+8]['champion_name'],data[holder+8]['team_position']],[data[holder+9]['champion_name'],data[holder+9]['team_position']]]
      
      if counter <= 4:
         data[i]['played_against'] = played_against_second_team
         data[i]['played_with'] = [champ for champ in played_against_first_team if champ[0] != data[i]['champion_name']]

      elif counter > 4:
         data[i]['played_against'] = played_against_first_team
         data[i]['played_with'] = [champ for champ in played_against_second_team if champ[0] != data[i]['champion_name']]

      counter += 1 

   return data

    
```

```python
# assign the team comps

player_stats = get_team_compositions(data)
```

```python
# swapping champion names with those positions

def normalize_positions(player_stats):
    
    for i in range(len(player_stats)):
        for champ in player_stats[i]['played_against']:
            for key, value in champions.items():
                if champ[0] in value:
                    champ[0] = key

    for i in range(len(player_stats)):
        for champ in player_stats[i]['played_with']:
            for key, value in champions.items():
                if champ[0] in value:
                    champ[0] = key
    return player_stats

player_stats = normalize_positions(player_stats)
```

```python
# i also need to tranform the data from numbers to actual names

item_dict = load_json('item.json')

simplified_items = {}

for id, info in item_dict['data'].items():
    simplified_items[id] = info['name']
```

```python
def assign_item_names(players):
    
    for player in players:
        for i, item in enumerate(player['items']):
            if item != 0:
                player['items'][i] = simplified_items[str(item)]
    return players

player_stats = assign_item_names(player_stats)
```

```python
# at this point we have clean data so now we need to create a dataframe 

import pandas as pd 

player_df = pd.DataFrame(player_stats)
```

```python
# i need to break each nested column into different columns

def break_champions(df, champs_against, champs_with):

    for index, row in df.iterrows():

        for position in df.loc[index, champs_against]:
            if position[1] == 'JUNGLE':
                df.loc[index, 'Enemy Jungler'] = position[0]
            elif position[1] == 'MIDDLE':
                df.loc[index, 'Enemy Middle'] = position[0]
            elif position[1] == 'BOTTOM':
                df.loc[index, 'Enemy Bottom'] = position[0]
            elif position[1] == 'TOP':
                df.loc[index, 'Enemy Top'] = position[0]
            elif position[1] == 'UTILITY':
                df.loc[index, 'Enemy Utility'] = position[0]

    for index, row in df.iterrows():

        for position in df.loc[index, champs_with]:
        
            if position[1] == 'JUNGLE':
                df.loc[index, 'Friendly Jungler'] = position[0]
            elif position[1] == 'MIDDLE':
                df.loc[index, 'Friendly Middle'] = position[0]
            elif position[1] == 'BOTTOM':
                df.loc[index, 'Friendly Bottom'] = position[0]
            elif position[1] == 'TOP':
                df.loc[index, 'Friendly Top'] = position[0]
            elif position[1] == 'UTILITY':
                df.loc[index, 'Friendly Utility'] = position[0]

    return df 
```

```python
# generally the concept is that i want to create a model that will take champion played as an input, position as an input and output 6 items

# let's first teach a katarina_mid model

katarina_mid = check[(check['champion_name'] == 'Katarina') & (check['team_position'] == 'MIDDLE')]

katarina_mid.dropna(inplace = True)
```

```python
# basically our prediction variable is a sequence - hence we could use a model that takes sequence as an input and outputs sequence as 
# an output. Maybe RNN then?
# but first we should transform the items data into something ready for a multilabel classification problem

from sklearn.preprocessing import MultiLabelBinarizer

def preprocess_labels(df, column):
    mlb = MultiLabelBinarizer()

    # Filter out zeros
    items_no_zeros = [[item for item in sublist if item != 0] for sublist in df[column].tolist()]

    # Fit and transform the labels
    labels = mlb.fit_transform(items_no_zeros)
    
    # Create a DataFrame of the labels
    labels_df = pd.DataFrame(labels, columns=mlb.classes_)

    # Reset indices
    df = df.reset_index(drop=True)
    labels_df = labels_df.reset_index(drop=True)
    
    # Concatenate original DataFrame with the labels DataFrame
    df = pd.concat([df, labels_df], axis=1)
    
    # Drop the original 'items' column
    # also let's drop champion name colimn and position name
    df = df.drop(columns=[column, 'champion_name', 'team_position'])

    
    return df, mlb.classes_

katarina_mid, item_columns = preprocess_labels(katarina_mid, 'items')
```

```python
# let's build an easy NN first to just see how it does

from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam

# Separate features and targets
X = katarina_mid.drop(columns=item_columns)  # Drops all item columns
y = katarina_mid[item_columns]  # Keeps only item columns

# Convert int type columns to float
X = X.astype(float)
y = y.astype(float)

# Define your model architecture
model = Sequential()
model.add(Dense(512, input_dim=X.shape[1], activation='relu'))  
model.add(Dropout(0.2)) # Adding dropout
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.2)) # Adding dropout
model.add(Dense(y.shape[1], activation='sigmoid'))

# Compile model with a custom learning rate
opt = Adam(learning_rate=0.001)
model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])

# Train the model
history = model.fit(X, y, epochs=50, batch_size=32, validation_split=0.2)
```
